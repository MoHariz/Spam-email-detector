# Splitting the data into training and testing set

```python
X = spam_df.drop(["is_spam"], axis=1)
Y = spam_df["is_spam"]

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# Using Naive Bayes algorithms without cross validation

```python
#Naive Bayes
from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score,confusion_matrix

# using Multinimoal Naive Bayes
Algo1 = MultinomialNB()
Algo1.fit(X_train,y_train)
result1 = Algo1.predict(X_test)


# using Gaussian Naive Bayes
Algo2 = GaussianNB()
Algo2.fit(X_train,y_train)
result2 = Algo2.predict(X_test)

print("Multinomial Naive Bayes confusion matrix")
print(confusion_matrix(y_test,result1))
print("Multinomial Naive Bayes score: ", Algo1.score(X_test,y_test))
print("")
print("Gaussian Naive Bayes confusion matrix")
print(confusion_matrix(y_test,result2))
print("Gaussian Naive Bayes score: ", Algo2.score(X_test,y_test))
```

# Using K Nearest Neighbors algorithms without cross validation

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train) 
result3 = neigh.predict(X_test)
KNN_cm = confusion_matrix(y_test,result3)

print("KNN predictions with K=3 and K-folds=10")
print(result3)
print("KNN confusion matrix with K=3 and K-folds=10")
print(KNN_cm)
print("KNN score: ", neigh.score(X_test,y_test))
print(KNN_cm[1,1])
```

# Applying cross-validation on K Nearest Neighbors

```python
from sklearn.model_selection import cross_val_score


# using 10 folds for cross validation
scores = cross_val_score(neigh, X, Y, cv=10, scoring='neg_mean_squared_error')
scores = scores * -1
print(scores)
```

```python
print("Accuracy: ",scores.mean())
```

# Finding the best K for K Nearest Neighbors

```python
# make a function that uses cross-validation with 10 folds and uses K Nearest Neighbors in a changing K value
def cv_KNN(data, x_train, y_train, K):
    X = data.drop(["is_spam"], axis=1)
    Y = data["is_spam"]
    neigh = KNeighborsClassifier(n_neighbors=K)
    neigh.fit(x_train, y_train)
    scores = cross_val_score(neigh, X, Y, cv=10, scoring='neg_mean_squared_error') * -1
    return(scores.mean())


Ks = []
RMSEs = []


# using the cross validation on KNN with changing K-value from 1 to 25
for i in range(25):
    num = i+1
    Ks.append(num)
    RMSEs.append(cv_KNN(spam_df, X_train, y_train, num))
    print("K=", num)
    print("RMSE: ", cv_KNN(spam_df, X_train, y_train, num))
    
    
```

```python
# graphing our result
plt.plot(Ks, RMSEs)
plt.xlabel("K")
plt.ylabel("RMSE")
plt.title("K-value vs. RMSE")

plt.show()
```

We tested the K Nearest Neighbors with 10 folds cross validations, and also tested it on different K value. From looking at this graph, we can see that the K value 5 gives the lowest root mean square error with an RMSE value of 0.21152835439141757.

# Applying cross-validation on Naive Bayes

```python
# Multinomial Naive Bayes
def cv_MNB(data, x_train, y_train, X_test, y_test):
    X = data.drop(["is_spam"], axis=1)
    Y = data["is_spam"]
    MNB = MultinomialNB()
    MNB.fit(x_train, y_train)
    y_pred = MNB.predict(X_test)
    scores = cross_val_score(MNB, X, Y, cv=10)
    recall = recall_score(y_test, y_pred, average="macro")
    precision = precision_score(y_test, y_pred, average="macro")
    f1 = f1_score(y_test, y_pred, average="macro")
    print("Multinomial -")
    print("Recall Score:", recall)
    print("Precrision Score:", precision)
    print("F-Score:", f1)
    print("CV-Score:", scores.mean())

# Gaussian Naive Bayes
def cv_GNB(data, x_train, y_train, X_test, y_test):
    X = data.drop(["is_spam"], axis=1)
    Y = data["is_spam"]
    GNB = GaussianNB()
    GNB.fit(x_train, y_train)
    y_pred = GNB.predict(X_test)
    scores = cross_val_score(GNB, X, Y, cv=10)
    recall = recall_score(y_test, y_pred, average="macro")
    precision = precision_score(y_test, y_pred, average="macro")
    f1 = f1_score(y_test, y_pred, average="macro")
    print("Gaussian: ")
    print("Recall Score:", recall)
    print("Precrision Score:", precision)
    print("F-Score:", f1)
    print("CV-Score:", scores.mean())

def cv_KNN2(data, x_train, y_train, K, X_test, y_test):
    X = data.drop(["is_spam"], axis=1)
    Y = data["is_spam"]
    neigh = KNeighborsClassifier(n_neighbors=K)
    neigh.fit(x_train, y_train)
    y_pred = neigh.predict(X_test)
    scores = cross_val_score(neigh, X, Y, cv=10)
    recall = recall_score(y_test, y_pred, average="macro")
    precision = precision_score(y_test, y_pred, average="macro")
    f1 = f1_score(y_test, y_pred, average="macro")
    print("KNN: ")
    print("Recall Score:", recall)
    print("Precrision Score:", precision)
    print("F-Score:", f1)
    print("CV-Score:", scores.mean())

cv_MNB(spam_df, X_train, y_train, X_test, y_test)
cv_GNB(spam_df, X_train, y_train, X_test, y_test)
cv_KNN2(spam_df, X_train, y_train, 5, X_test, y_test)
```
